# In many ways, the most effiecient sorting alogorithm is the Quick sort algorithm.
## to do a quick sort, pick one of the values in the array at random and move all values greater than it above it and all values less than it below it. The initial value is called a pivot. It is conventianal to pick the far right element
## you then recursively pick values in the upper and lower areas of the array and sort them similiarly

## the effeicency of quick sort is very complicated
## it works by splitting the array in half evry time. so the worst case will be if you can split the array in half, such as when the pivot is already in the right place. ultimately, the worst case of quick sort is just niave sort, $O(n^{2})$
### On average however, it's complexity is O(nlog(n))
# don't use quick sort if something is nearly sorted
### you can optimize by running both halfs at the samet time as well as taking a median of group on the end as the pivot

```python
"""Implement quick sort in Python.
Input a list.
Output a sorted list."""


# Quick sort in Python

# function to find the partition position
def partition(array, low, high):

    # choose the rightmost element as pivot
    pivot = array[high]

    # pointer for greater element
    i = low - 1

    # traverse through all elements
    # compare each element with pivot
    for j in range(low, high):
        if array[j] <= pivot:
              # if element smaller than pivot is found
              # swap it with the greater element pointed by i
              i = i + 1

              # swapping element at i with element at j
              (array[i], array[j]) = (array[j], array[i])

    # swap the pivot element with the greater element specified by i
    (array[i + 1], array[high]) = (array[high], array[i + 1])

  # return the position from where partition is done
    return i + 1

# function to perform quicksort
def quickSort(array, low=0, high =-1):
    if high ==-1:
        high =len(array)-1
    if low < high:

        # find pivot element such that
        # element smaller than pivot are on the left
        # element greater than pivot are on the right
        pi = partition(array, low, high)

        # recursive call on the left of pivot
        quickSort(array, low, pi - 1)

        # recursive call on the right of pivot
        quickSort(array, pi + 1, high)
    return array

test = [21, 4, 1, 3, 9, 20, 25, 6, 21, 14]
print(quickSort(test))


```

# Hash functions
## takes in a value and spits out a hash value: a coded version of the value
## A hash table is a DS that allows for the very fast retrieval no matter how much data there is.
## key = "bob" ,  address = "index number"
## Every element is indexed based on it's key. To make fast retrieval, relate the key to the index number(address) so that you can plug in the key you're looking for and it's address will be revealed.
## A hashmap is a hash table of key-value pairs. Every element can be an object, with many attributes and one hash value
# A python dictionary is a hash-map
### address = key(Mod(num of elements))
* ### for alphanumeric keys, address can be sum of ASCII val in key / num_of_addresses = remainder
* ### Folding method, break up a long key into many 2 dig num and sum/constant = remainder is address

## so when two keys create them same address, its a collision and no single solution for this.
### when you have a key with an already used address, putting it at an open address is called Open-addressing. Using a linear search to find this open address is called **Linear Probing**. However, this makes it hard to find items also, invloving a linear search if it's not at the calculated address.
### The more open addresses in the array, the less likely a colission. This percentage of openess is called the **Load Factor**. As long as it's low, Open addressing + linear probing should work well.
### $Number of Entries / Number of Buckets$ The purpose of a load factor is to give us a sense of how "full" a hash table is. For example, if we're trying to store 10 values in a hash table with 1000 buckets, the load factor would be 0.01, and the majority of buckets in the table will be empty. We end up wasting memory by having so many empty buckets, so we may want to rehash, or come up with a new hash function with less buckets. We can use our load factor as an indicator for when to rehash—as the load factor approaches 0, the more empty, or sparse, our hash table is. On the flip side, the closer our load factor is to 1 (meaning the number of values equals the number of buckets), the better it would be for us to rehash and add more buckets. Any table with a load value greater than 1 is guaranteed to have collisions. 

## Another solution is called chaining or closed addressing.
### Every repeat becomes the next element in a linked-list headed by the first element with that address. To search this, just calculate the address and use a standard linked list traversal. If the load factor is low, it may be more effiecient to just use linear probing, as ther well be very few collisisions
## Some other collision resolution ideas:
* ### plus three rehash: probe multiples of three forward in the search for openess
* ### quadratic probing: search at the (failed attempts)^2 address for openess
* ### double hasheing: apply a second hash function to the collided keys

## if you know all of the keys in advance, it's theoretically posible to come up with a perfect hash function in advance, no collisions

* ### one approach is Depth-first-search (DFS): Priority on visiting the child nodes. 
#### Pre-order = start at root and always left your way down the tree checking each node as you hit it, until you hit a leaf, then go to left most, upper most, unchecked leaf . 
#### In-order = start with leftmost leaf and go up, checking off until you reach a parent with another child, check it off and then leftmost down to the bottom leaf of that child. This is called in order because we are always hitting the leftmost node on the X-plane  
#### Post-Order =start with leftmost leaf and go up, checking off until you reach a parent with another child, DON'T check it off, instead leftmost down to the bottom leaf of that child. when you checked all children of a parent, check the parent. This was done right if the root is the last node checked
* ### Bredth-First-search(BFS): priority on visiting the entire level. Level-order. Start on the left, at the root

## Binary Search Tree (BST)
### a subtype of Binary trees. They are value ordered, every value to the left of a node is smaller than the parent and every value to the right is larger.
### This make search, insert much quicker, delete is the same.
### The search time of a BST is O(log(n)) AKA the height of the tree.
### An unbalanced BST is basically just a chain of right children, You have to keep looking higher to find what your looking for, could be as long as O(n)

```python
class Node(object):
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BST(object):
    def __init__(self, root):
        self.root = Node(root)

    def insert(self, new_val):
        return self.inserter(self.root, new_val)

    def search(self, find_val):
        return self.searcher(self.root, find_val)
    
    def inserter(self, start, new_val):
        """Helper method - use this to create a 
        recursive search solution."""
        if new_val == start.value:
            return "already in the tree"
        
        if new_val < start.value:
            #print(new_val,'goes left of ', start.value)
            if start.left:
                self.inserter(start.left, new_val)
            else:
                start.left = Node(new_val)
                #print(start.left.value, 'inserted to the left of', start.value)
                return start

        if new_val > start.value:
            #print(new_val,'goes right of ', start.value)
            if start.right:
                self.inserter(start.right, new_val)
            else:
                start.right = Node(new_val)
                #print(start.right.value, 'inserted to the right of', start.value)
                return start
        
    def searcher(self, start, find_val):
        """Helper method - use this to create a 
        recursive search solution."""
        if find_val == start.value: # If start value is the one we're looking for
            return True

        if find_val < start.value:  # If start value is greater than the one we're looking for
            if start.left:             # And there's a lower value
                return self.searcher(start.left, find_val) # pause this function call and call the function on that lower value

        if find_val > start.value: # If start value is less than the one we're looking for
            if start.right: # And there's a greater value
                return self.searcher(start.right, find_val)  # pause this function call and call the function on that greater value                    
        return False
    
# Set up tree
tree = BST(4)

# Insert elements
tree.insert(2)
tree.insert(1)
tree.insert(3)
tree.insert(5)

# Check search
print(tree.root.value)
print(tree.root.left.value)
print(tree.root.left.left.value)
print(tree.root.right.value)
# Should be True
print(tree.search(4))
# Should be False
print(tree.search(6))

print(tree.search(5))

print(tree.search(3))
```

## Depth first search (DFS): we follow one path as far as it will go
### In DFS: search and traversal are basically the same thing. Begin with any node, and mark it as "Seen" in the stack.
### Follow an edge to a node and mark that one as "Seen" on the stack
### If you hit a node that you've already seen, go back and try a different edge. If there are no more edges on that node, pop the current node off the stack and look at the previous node
### this continues until you've popped everthing off the stack or you've found the Node you were looking for.

### the efficiency of this algorithm is O(|Edges| + |Verteses|), technically it's 2 x O(|Edges| + |Verteses|) because we actually visit every thing going forward and then going back, but we approx it

## Bredth first search (BFS): we look at all nodes adjacent to one before moving on to the next level.
### In BFS: Start with a node and follow an edge to a node. Then put that adj node in a FIFO Queue.
### proceed to the next adjacent node to the starting node and do the same add to a FIFO Queue.
### When we've explored all of the edges of the starting node, remove the FO node from the Queue and begin following that node's edges
### the efficiency of this algorithm is O(|Edges| + |Verteses|), technically it's 2 x O(|Edges| + |Verteses|) because we actually visit every thing going forward and then going back, but we approx it

## Notable paths:
* ### Eulerian path: A path that travels through every edge at least once. Start from one node and keep traversing until you've used every edge. 
#### If it's an Eulerian cycle, then you use every edge only once and end up at the node you started with. A graph can only have this if all the nodes have an even degree (number of edges).
#### Start at any vertex and follow edges until you return to that vertex. Then, start again from a node with unseen edges and do the same with those unseen edges. Add of the loops at their intersections to get a path, by putting the common element last in the first loop and first in the second loop. Effieincy is O(edges)

* ### Hamiltonian Path: Must go through every edge at least once, and must start and end with the same node

## The Shortest path problem: the sum of the edges is a small as possible. 
### Unweighted graph: solution is just a BFS
### weighted non directional: Dijkstra's Algorithm. Start with a node and add it to a priority queue with a and edge weight of 0 (lowest gets priority bc shortest path). Then, add the edge weight of all the adjacent edges to the priority queue. Then, beggining from the lowest edge wieght in the queue, remove all of the others and add all of the edges weights which surround that lowest edge weight.  

## The Knapsack problem: we have a bag with a limited weight cappacity, and more items than can possibly fit into it. Each item has a weigt and a value
* ### Brute force sol: $O(2^{n})$
* ### Max Value For Min Weight: 
#### Create a Set with all possible weights and they're ideal value. Take the lowest weight object and set it's value as the ideal for all weights 
#### Then take the next lowest weight object and if it's value is greater than then the prev or if the sum of two dif lower weight, set it's value as the ideal for all wieghts. 
#### Efficiency is $O(n*W)$, psuedo-polynomial time
### This shows the beuaty of Dynamic programming
#### Base case, with a lookup table with ideal solutions for smaller problems called memoizations
#### If a problem can be solved by trying every combination, look for the dynamic subproblems and memoize them
