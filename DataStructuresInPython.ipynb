# In many ways, the most effiecient sorting alogorithm is the Quick sort algorithm.
## to do a quick sort, pick one of the values in the array at random and move all values greater than it above it and all values less than it below it. The initial value is called a pivot. It is conventianal to pick the far right element
## you then recursively pick values in the upper and lower areas of the array and sort them similiarly

## the effeicency of quick sort is very complicated
## it works by splitting the array in half evry time. so the worst case will be if you can split the array in half, such as when the pivot is already in the right place. ultimately, the worst case of quick sort is just niave sort, $O(n^{2})$
### On average however, it's complexity is O(nlog(n))
# don't use quick sort if something is nearly sorted
### you can optimize by running both halfs at the samet time as well as taking a median of group on the end as the pivot

```python
"""Implement quick sort in Python.
Input a list.
Output a sorted list."""


# Quick sort in Python

# function to find the partition position
def partition(array, low, high):

    # choose the rightmost element as pivot
    pivot = array[high]

    # pointer for greater element
    i = low - 1

    # traverse through all elements
    # compare each element with pivot
    for j in range(low, high):
        if array[j] <= pivot:
              # if element smaller than pivot is found
              # swap it with the greater element pointed by i
              i = i + 1

              # swapping element at i with element at j
              (array[i], array[j]) = (array[j], array[i])

    # swap the pivot element with the greater element specified by i
    (array[i + 1], array[high]) = (array[high], array[i + 1])

  # return the position from where partition is done
    return i + 1

# function to perform quicksort
def quickSort(array, low=0, high =-1):
    if high ==-1:
        high =len(array)-1
    if low < high:

        # find pivot element such that
        # element smaller than pivot are on the left
        # element greater than pivot are on the right
        pi = partition(array, low, high)

        # recursive call on the left of pivot
        quickSort(array, low, pi - 1)

        # recursive call on the right of pivot
        quickSort(array, pi + 1, high)
    return array

test = [21, 4, 1, 3, 9, 20, 25, 6, 21, 14]
print(quickSort(test))


```

# Hash functions
## takes in a value and spits out a hash value: a coded version of the value
## A hash table is a DS that allows for the very fast retrieval no matter how much data there is.
## key = "bob" ,  address = "index number"
## Every element is indexed based on it's key. To make fast retrieval, relate the key to the index number(address) so that you can plug in the key you're looking for and it's address will be revealed.
## A hashmap is a hash table of key-value pairs. Every element can be an object, with many attributes and one hash value
# A python dictionary is a hash-map
### address = key(Mod(num of elements))
* ### for alphanumeric keys, address can be sum of ASCII val in key / num_of_addresses = remainder
* ### Folding method, break up a long key into many 2 dig num and sum/constant = remainder is address

## so when two keys create them same address, its a collision and no single solution for this.
### when you have a key with an already used address, putting it at an open address is called Open-addressing. Using a linear search to find this open address is called **Linear Probing**. However, this makes it hard to find items also, invloving a linear search if it's not at the calculated address.
### The more open addresses in the array, the less likely a colission. This percentage of openess is called the **Load Factor**. As long as it's low, Open addressing + linear probing should work well.
### $Number of Entries / Number of Buckets$ The purpose of a load factor is to give us a sense of how "full" a hash table is. For example, if we're trying to store 10 values in a hash table with 1000 buckets, the load factor would be 0.01, and the majority of buckets in the table will be empty. We end up wasting memory by having so many empty buckets, so we may want to rehash, or come up with a new hash function with less buckets. We can use our load factor as an indicator for when to rehashâ€”as the load factor approaches 0, the more empty, or sparse, our hash table is. On the flip side, the closer our load factor is to 1 (meaning the number of values equals the number of buckets), the better it would be for us to rehash and add more buckets. Any table with a load value greater than 1 is guaranteed to have collisions. 

## Another solution is called chaining or closed addressing.
### Every repeat becomes the next element in a linked-list headed by the first element with that address. To search this, just calculate the address and use a standard linked list traversal. If the load factor is low, it may be more effiecient to just use linear probing, as ther well be very few collisisions
## Some other collision resolution ideas:
* ### plus three rehash: probe multiples of three forward in the search for openess
* ### quadratic probing: search at the (failed attempts)^2 address for openess
* ### double hasheing: apply a second hash function to the collided keys

## if you know all of the keys in advance, it's theoretically posible to come up with a perfect hash function in advance, no collisions

* ## You can describe a tree in terms of levels: how many moves it takes to reach the root+1
* ## we can use the parent child language, only one parent per node
* ## nodes at the edge which have no children are called leaves (external nodes)
* ## a connection is an edge and group of connections is a path
* ## depth of a node is the number of edges it takes to get to the root 
### When we look at list data-structures, traversing was pretty straight. Trees are not straight so how do we traverse them?
* ### one approach is Depth-first-search (DFS): Priority on visiting the child nodes. Pre-order = total left to right down the tree. In-order = start with leftmost leaf and go up. Post-Order = check your way to the leftmost leaf and then go up 
* ### Bredth-First-search(BFS): priority on visiting the entire level. Level-order
